# **Logistic Regression**

```python
# Load your data
dff = pd.read_csv('/content/polynomial_regression_data.csv')

# Rename columns
dff.columns = ['Year', 'Hard_Disc_Capacity_TB', 'Processor_Speed_GHz', 'Price_of_Computer_USD']

dff.head()
```

```python
bins = [0, 5000, 8000, float('inf')]
labels = ['low', 'medium', 'high']

dff['budget'] = pd.cut(dff['Price_of_Computer_USD'], bins=bins, labels=labels, right=False)

dff.head()
```

```python
dff['budget_encoded'] = dff['budget'].map({'low': 0, 'medium': 1, 'high': 2})
dff.head()
```

```python
X = dff.iloc[:, :-2]
np.array(X)
```

```python
y = dff['budget_encoded']
y
```

```python
x=np.array(X)
y=np.array(y).reshape(len(y),1)
print(x.shape,y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.4, random_state=1)
X_test, X_val, Y_test, Y_val = train_test_split(X_test, Y_test, test_size=0.5, random_state=1)

print('shape of training set:',X_train.shape,Y_train.shape)
print('shape of validation set:',X_val.shape,Y_val.shape)
print('shape of test set:',X_test.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, roc_auc_score

logistic_model = LogisticRegression(multi_class='ovr', max_iter=1000)
logistic_model.fit(X_train, Y_train.ravel())

Y_pred = logistic_model.predict(X_train)
Y_prob = logistic_model.predict_proba(X_train)

conf_matrix = confusion_matrix(Y_train, Y_pred)
accuracy = accuracy_score(Y_train, Y_pred)
recall = recall_score(Y_train, Y_pred, average='weighted')
precision = precision_score(Y_train, Y_pred, average='weighted')

roc_auc = roc_auc_score(Y_train, Y_prob, multi_class='ovr')

print('Confusion Matrix:\n', conf_matrix)
print('Accuracy:', accuracy)
print('Recall:', recall)
print('Precision:', precision)
print('ROC AUC:', roc_auc)
```

```python
Y_pred = logistic_model.predict(X_test)
Y_prob = logistic_model.predict_proba(X_test)

conf_matrix = confusion_matrix(Y_test, Y_pred)
accuracy = accuracy_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred, average='weighted')
precision = precision_score(Y_test, Y_pred, average='weighted')

roc_auc = roc_auc_score(Y_test, Y_prob, multi_class='ovr')

print('Confusion Matrix:\n', conf_matrix)
print('Accuracy:', accuracy)
print('Recall:', recall)
print('Precision:', precision)
print('ROC AUC:', roc_auc)
```

```python
Y_pred = logistic_model.predict(X_val)
Y_prob = logistic_model.predict_proba(X_val)

conf_matrix = confusion_matrix(Y_val, Y_pred)
accuracy = accuracy_score(Y_val, Y_pred)
recall = recall_score(Y_val, Y_pred, average='weighted')
precision = precision_score(Y_val, Y_pred, average='weighted')

roc_auc = roc_auc_score(Y_val, Y_prob, multi_class='ovr')

print('Confusion Matrix:\n', conf_matrix)
print('Accuracy:', accuracy)
print('Recall:', recall)
print('Precision:', precision)
print('ROC AUC:', roc_auc)
```
